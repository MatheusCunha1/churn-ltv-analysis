## Dados Categóricos

```python
#card_brand
duckdb.sql(f"""
    SELECT 
        card_brand, 
        COUNT(*) total
    FROM read_csv_auto('{base}/data/bronze/cards_data.csv')
    GROUP BY card_brand
    ORDER BY 2 DESC
    """).show()

#card_type
duckdb.sql(f"""
    SELECT 
        card_type    , 
        COUNT(*) total
    FROM read_csv_auto('{base}/data/bronze/cards_data.csv')
    GROUP BY card_type
    ORDER BY 2 DESC
    """).show()

#has_chip             
duckdb.sql(f"""
    SELECT 
        has_chip, 
        COUNT(*) total
    FROM read_csv_auto('{base}/data/bronze/cards_data.csv')
    GROUP BY has_chip
    ORDER BY 2 DESC
    """).show()

#num_cards_issued             
duckdb.sql(f"""
    SELECT 
        num_cards_issued, 
        COUNT(*) total
    FROM read_csv_auto('{base}/data/bronze/cards_data.csv')
    GROUP BY num_cards_issued
    ORDER BY 2 DESC
    """).show()

#year_pin_last_changed             
duckdb.sql(f"""
    SELECT 
        year_pin_last_changed, 
        COUNT(*) total
    FROM read_csv_auto('{base}/data/bronze/cards_data.csv')
    GROUP BY year_pin_last_changed
    ORDER BY 2 DESC
    """).show()



#card_on_dark_web
duckdb.sql(f"""
    SELECT 
        card_on_dark_web, 
        COUNT(*) total
    FROM read_csv_auto('{base}/data/bronze/cards_data.csv')
    GROUP BY card_on_dark_web
    ORDER BY 2 DESC
    """).show()
```
